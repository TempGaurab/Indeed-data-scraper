```python
def removal(path):
    with open(path, 'r') as file:
        content = file.read()
        content = f" {content} "
        content = content.replace("'", "")
    return content
```

```python
path = "temp.txt"
```

```python
content = removal(path)
content
```

```python
with open(path, 'w') as file:
    file.write(content)
```
