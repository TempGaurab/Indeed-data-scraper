```python
import os
import shutil
import glob
```

```python
def move_txt_files(source_folder, old_folder):
    # Ensure the old_folder exists
    if not os.path.exists(old_folder):
        os.makedirs(old_folder)

    # List all .txt files in the source_folder
    file_list = [f for f in os.listdir(source_folder) if f.endswith('.txt')]
    
    for file_name in file_list:
        source_file_path = os.path.join(source_folder, file_name)
        old_file_path = os.path.join(old_folder, file_name)
        
        # Move the file to the old_folder
        shutil.move(source_file_path, old_file_path)
        print(f'Moved {file_name} to {old_folder}')

# Example usage
source_folder = '.'
old_folder = 'old_txt'
move_txt_files(source_folder, old_folder)
```
