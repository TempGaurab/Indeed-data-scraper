```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
```

```python
# Set up Chrome options
options = Options()
options.headless = True  # Run in headless mode
```

```python
chrome_driver_path = r"C:\Users\gauur\projects\WEB SCRAPE\Indeed-Job-Scraping\chromedriver.exe"
```

```python
service = Service(chrome_driver_path)
service
```

```python
driver = webdriver.Chrome(service=service, options=options)
driver
```

```python
# Visit the URL
url = "https://www.indeed.com/jobs?q=artificial+intelligence&l=United+States"
driver.get(url)
```

```python
page_source = driver.page_source
page_source
```

```python
value = page_source
value
```

```python
with open("txt_files/indeed_job_search_results.txt", "w", encoding="utf-8") as file:
    file.write(value)
```

```python
driver.quit()
```
