```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
import os
import glob
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
base_url = "https://www.linkedin.com/jobs/search/?currentJobId=3982384918&distance=25&geoId=103644278&keywords=artificial%20intelligence&origin=JOB_SEARCH_PAGE_SEARCH_BUTTON&refresh=true&start="
```

```python
def getjob_ids(page_source):
    job_ids = []
    soup = BeautifulSoup(page_source, 'html.parser')
    job_listing = soup.findAll('li', class_='ember-view')
    for job in job_listing:
        job_id = job.get('data-occludable-job-id')
        if job_id:
            job_ids.append(job_id)
    return job_ids
```

```python
# Generate URLs with different 'start' values
start_values = range(0, 976, 25)  # From 0 to 100, stepping by 10
urls = [base_url + str(start) for start in start_values]
urls[len(urls)-1]
```

```python
# Set up Chrome options
options = Options()
options.headless = True  # Run in headless mode
```

```python
chrome_driver_path = r"C:\Users\gauur\projects\WEB SCRAPE\Indeed-Job-Scraping\chromedriver.exe"
```

```python
service = Service(chrome_driver_path)
service
```

```python
driver = webdriver.Chrome(service=service, options=options)
driver
```

```python
all_job_ids = []
for i, url in enumerate(urls):
    try:
        # Navigate to the URL
        driver.get(url)
        # Wait for the page to load
        time.sleep(10)
        
        # Get the page source HTML
        page_source = driver.page_source
        
        # Get job IDs from the page source
        job_ids = getjob_ids(page_source)
        
        # Store or process the extracted job IDs
        all_job_ids.extend(job_ids)
    except Exception as e:
        print(f"Error processing URL {url}: {e}")

# Close the driver
driver.quit()
```
